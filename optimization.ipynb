# Общие параметры

```python
from IPython.display import display, Math
from math import sqrt

# The function to be changed
def F(x):
    return sqrt(x**2-4*x+10)/5
    #return 2 * (x-12)*(x-12) + 3

def dm(eq):
    display(Math(eq))
eps = 1
```

# Алгоритм золотого сечения

## Начальные параметры

```python
tau = 0.618
a_0 = 0
b_0 = 13
```

## Решение

```python
r = 1
a = a_0
b = b_0
while (b - a) > eps:
    dm(f"\\textbf{{Итерация}}\\ {r}")
    dm(f"ТИН_{r} = [a^{r}, b^{r}] = [{a},{b}]")
    dm(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}")

    x1 = round(b - (b - a) * tau, 2)
    x2 = round(a + (b - a) * tau, 2)
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)

    dm(f"x_1^{{{r}}} = b^{{{r}}} - (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {b} - ({b} - {a}) \\cdot {tau} = {x1}")
    dm(f"x_2^{{{r}}} = a^{{{r}}} + (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {a} + ({b} - {a}) \\cdot {tau} = {x2}")

    dm(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}")
    dm(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}")

    if F1 < F2:
        dm(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}")
        b = x2
    else:
        dm(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}")
        a = x1

    r += 1

    dm(f"ТИН_{r} = [{a},{b}]")
    dm(f"|ТИН_{r}| = {b} - {a} = {round(b-a, 2)} {'>' if (b-a) > eps else '<'} {eps}")

print("\n")
F_ans = min(F(a), F(b))
dm(f"\\Phi({a}) = {F(a):.2f}")
dm(f"\\Phi({b}) = {F(b):.2f}")
dm(f"\\Phi(x^{{*}}) = min(\\Phi({a}), \\Phi({b})) = {F_ans:.2f}")
```

# Алгоритм дихотомического поиска

## Начальные параметры

```python
delta_x = 0.2
```

## Решение

```python
r = 1
a = a_0
b = b_0
while (b - a) > eps:
    dm(f"\\textbf{{Итерация}}\\ {r}")
    dm(f"ТИН_{r} = [a^{r}, b^{r}] = [{a},{b}]")
    dm(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}\n")
    x0 = round((a + b) / 2, 2)
    x1 = round(x0 - delta_x / 2, 2)
    x2 = round(x0 + delta_x / 2, 2)
    
    dm(r"x_0^{" + str(r) + r"} = \frac{a^{" + str(r) + r"} + b^{" + str(r) + r"}}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x0))
    dm(r"x_1^{" + str(r) + r"} = x_0^{" + str(r) + r"} - \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x1))
    dm(r"x_2^{" + str(r) + r"} = x_0^{" + str(r) + r"} + \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x2))
    
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)
    
    dm(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}")
    dm(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}")
    
    if F1 < F2:
        dm(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}")
        b = x2
    else:
        dm(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}")
        a = x1
    r += 1

    dm(f"ТИН_{r} = [{a},{b}]")
    dm(f"|ТИН_{r}| = {b} - {a} = {round(b-a, 2)} {'>' if (b-a) > eps else '<'} {eps}")
    
print("\n")
F_ans = min(F(a), F(b))
dm(f"\\Phi({a}) = {F(a):.2f}")
dm(f"\\Phi({b}) = {F(b):.2f}")
dm(f"\\Phi(x^{{*}}) = min(\\Phi({a}), \\Phi({b})) = {F_ans:.2f}")
```

# Метод квадратичной аппроксимации

## Начальные параметры

Надо условится, что у нас уже имеется функция (с этим проблем нет) и три точки

```python
x1 = 0
x2 = 2
x3 = 5
```

## Решение

```python
a = x1
b = x3

r = 1

dm(f"x_1 = {a},\\quad x_2 = {x2},\\quad x_3 = {b}")

F1 = round(F(x1), 2)
F2 = round(F(x2), 2)
F3 = round(F(x3), 2)

dm(f"\\textbf{{Не забудьте расписать численные выражения Ф}}")
dm(f"\\Phi(x_1) = \\Phi({x1}) \\approx {F1}")
dm(f"\\Phi(x_2) = \\Phi({x2}) \\approx {F2}")
dm(f"\\Phi(x_3) = \\Phi({x3}) \\approx {F3}")

print("\n")

dm("\\textbf{{Найдём коэффициенты аппроксимирующего полинома}}")
dm("q(x) = a_0 + a_1(x-x_1) + a_2(x-x_1)(x-x_2)")

a0 = F1
a1 = round((F2 - F1) / (x2 - x1),2)
a2 = round(1/(x3 - x2) * ((F3 - F1)/(x3 - x1) - (F2 - F1)/(x2 - x1)), 2)
x_tilde = round((x2 + x1)/2 - a1/(2 * a2), 2)

print("\n")
dm(f"q(x_1) = a_0 + a_1(x_1-x_1) + a_2(x_1-x_1)(x_1-x_2) = a_0 = \\Phi(x_1) = {F1}")
dm(f"a_0 = {a0}")
print("\n")

dm(f"q(x_2) = a_0 + a_1(x_2-x_1) + a_2(x_2-x_1)(x_2-x_2) = a_0 + a_1(x_2-x_1) = {a0} + a_1({x2} - {x1}) = {a0} {x2 - x1:+.2f}a_1 = \\Phi(x_2) = {F2}")
dm(f"a_1 = \\frac{{{F2} {'-' if (a0 > 0) else '+'} {abs(a0)}}}{{{round(x2 - x1, 2)}}} = {a1}")
print("\n")

t = round(a0 + a1 * (x3 - x1), 2)
dm(f"q(x_3) = a_0 + a_1(x_3-x_1) + a_2(x_3-x_1)(x_3-x_2) = {a0} {a1:+}\\cdot({x3} - {x1}) + a_2({x3} - {x1})({x3} - {x2}) = {a0} {t:+} "
  f"{(x3 - x1)*(x3 - x2):+.2f}a_2 = \\Phi(x_3) = {F3}")
dm(f"a_2 = \\frac{{{F3} {'-' if ((a0 + a1 * (x3 - x1)) > 0) else '+'} {abs(t)}}}{{{round((x3-x1)*(x3-x2), 2)}}} = {a2}")
print("\n")

dm("\\textbf{{Таким образом, аппроксимирующий полином имеет вид:}}")
dm(f"q(x) = {a0} {a1:+}(x {'-' if (x1 > 0) else '+'}{abs(x1)}) {a2:+}(x {'-' if (x1 >= 0) else '+'}{abs(x1)})(x {'-' if (x2 > 0) else '+'}{abs(x2)})")
print("\n")

x_part = 2 * a2
c_part = a1 - a2 * x1 - a2 * x2
dm("\\textbf{Выражение для первой производной:}")
dm(f"q'(x) = {a1:.2f} + {a2:.2f} \\cdot (2x - {x1:.2f} - {x2:.2f}) = {x_part}x {c_part:+}")

dm("\\textbf{Экстремуму аппроксимирующей функции соответствует точка } \\tilde{x}, \\text{ обращающая производную } q'(x) \\text{ в ноль.}")
dm(f"{x_part}x {c_part:+} = 0")
dm(f"{x_part}x = {'-' if (c_part > 0) else '+'}{abs(c_part)}")
dm(f"\\tilde{{x}} ={x_tilde} ")

q_tilde = round(a0 + a1 * (x_tilde - x1) + a2 * (x_tilde - x1)*(x_tilde - x2), 4)
F_tilde = round(F(x_tilde), 4)
dm(f"q(\\tilde{{x}}) = {a0} {a1:+} \\cdot ({x_tilde} {'-' if (x1 >= 0) else '+'}{abs(x1)}) {a2:+} \\cdot ({x_tilde} {'-' if (x1 >= 0) else '+'}{abs(x1)})({x_tilde} {'-' if (x2 >= 0) else '+'}{abs(x2)}) = {q_tilde}")
print("\n")

dm("\\textbf{Значение целевой функции } \\Phi(x) \\textbf{ в точке } \\tilde{x} \\textbf{ будет равно:}")
dm(f"\\Phi(\\tilde{{x}}) = \\Phi({x_tilde}) = {F_tilde}") 

error = abs(q_tilde - F_tilde)
dm("\\textbf{Найдём погрешность аппроксимации } \\epsilon_{f}:")
dm(f"\\epsilon_f = | q(\\tilde{{x}}) - \\Phi(\\tilde{{x}}) | = {q_tilde} {'-' if (F_tilde > 0) else '+'}{abs(F_tilde)} = {error:.4f}")

dm(f"x^{{*}} \\approx \\tilde{{x}} = {x_tilde}")
dm(f"\\Phi(x^{{*}}) \\approx \\Phi(\\tilde{{x}}) = {F_tilde}")
```

# Алгоритм Фибоначчи

## n-ый элемент ряда Фибоначчи

```python
def n_fib(n):
    τ = (sqrt(5)-1)/2
    return round(((1/τ)**(n+1)-((-τ)**(n+1)))/sqrt(5))
```

## Решение

```python
a=a_0
b=b_0
N = 1
while (b-a)/n_fib(N) > eps:
    N += 1
dm(rf"N={N}")

print("Формулы:")
dm(r"x_{1}^{r}=a^r+|ТИН_r|\frac{i_{N-1-r}}{i_{N+1-r}}")
dm(r"x_{2}^{r}=a^r+|ТИН_r|\frac{i_{N-r}}{i_{N+1-r}}")
print("\n\n") # 3 empty lines for identation

for r in range(1, N): # [1; N-1]
    print(rf"Итерация {r} из {N-1}: ")
    fib1 = n_fib(N-1-r) / n_fib(N+1-r)
    fib2 = n_fib(N-r) / n_fib(N+1-r)
    x1_r = a + abs(b - a) * fib1
    x2_r = a + abs(b - a) * fib2
    Phi1 = F(x1_r)
    Phi2 = F(x2_r)
    
    dm(rf"a_{r} = {a};\ b_{r}={b}")
    dm(rf"\frac{{i_{N-1-r}}}{{i_{N+1-r}}}=\frac{{{n_fib(N-1-r)}}}{{{n_fib(N+1-r)}}};\ \frac{{i_{N-r}}}{{i_{N+1-r}}}=\frac{{{n_fib(N-r)}}}{{{n_fib(N+1-r)}}}")
    dm(rf"x_{1}^{r}={a}+|{b}-{a}|\frac{{i_{N-1-r}}}{{i_{N+1-r}}}={x1_r}")
    dm(rf"x_{2}^{r}={a}+|{b}-{a}|\frac{{i_{N-r}}}{{i_{N+1-r}}}={x2_r}")
    dm(rf"\Phi({{x_1^{r}}})={Phi1}")
    dm(rf"\Phi({{x_2^{r}}})={Phi2}")

    if r != N - 1:
        if Phi1 < Phi2:
            b = x2_r
            dm(rf"\Phi({{x_1^{r}}}) < \Phi({{x_2^{r}}}) \Rightarrow b_{r+1} = x_2^{r} = {x2_r}")
        else:
            a = x1_r
            dm(rf"\Phi({{x_1^{r}}}) > \Phi({{x_2^{r}}} )\Rightarrow a_{r+1} = x_1^{r} = {x1_r}")
    
    print("\n") # 2 empty lines for identation

assert x1_r == x2_r
print("Этап 2")
r += 1

delta_x = (b - a)/20
x2_r = x1_r + delta_x
Phi1 = F(x1_r)
Phi2 = F(x2_r)

dm(rf"a_{r} = {a};\ b_{r}={b}")
dm(rf"\text{{Пусть }}\delta_x=\frac{{|ТИН_r|}}{{20}}={delta_x}")
dm(rf"x_{1}^{r} = {x1_r}")
dm(rf"x_{2}^{r} = x_{1}^{r-1} + \delta_x = {x2_r}")
dm(rf"\Phi({{x_1^{r}}})={Phi1}")
dm(rf"\Phi({{x_2^{r}}})={Phi2}")

if Phi1 < Phi2:
    b = x1_r
    dm(rf"\Phi({{x_1^{r}}}) < \Phi({{x_2^{r}}}) \Rightarrow b_{r} = x_1^{r} = {x1_r}")
else:
    a = x1_r
    dm(rf"\Phi({{x_1^{r}}}) > \Phi({{x_2^{r}}}) \Rightarrow a_{r} = x_1^{r} = {x1_r}")
dm(rf"a_{r} = {a};\ b_{r}={b}")
```
