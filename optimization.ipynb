```python
from IPython.display import display, Math

# Change function
def F(x):
    return 2 * (x-12)*(x-12) + 3
eps = 1.0
a = 0
b = 13
```

# Алгоритм золотого сечения

Enter your $\tau$

```python
tau = 0.618
```

```python
r = 1
while (b - a) > eps:
    display(Math(f"\\textbf{{Итерация}}\\ {r}"))
    display(Math(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}"))

    x1 = round(b - (b - a) * tau, 2)
    x2 = round(a + (b - a) * tau, 2)
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)

    display(Math(f"x_1^{{{r}}} = b^{{{r}}} - (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {b} - ({b} - {a}) \\cdot {tau} = {x1}"))
    display(Math(f"x_2^{{{r}}} = a^{{{r}}} + (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {a} + ({b} - {a}) \\cdot {tau} = {x2}"))

    display(Math(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}"))
    display(Math(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}"))

    if F1 < F2:
        display(Math(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}"))
        b = x2
    else:
        display(Math(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}"))
        a = x1

    r += 1

display(Math(f"\\textbf{{Ответ:}}\\quad a = {a},\\quad b = {b}"))
```

# Алгоритм дихотомического поиска

Enter your $\delta_x$

```python
delta_x = 0.2
```

```python
r = 1
while (b - a) > eps:
    display(Math(f"\\textbf{{Итерация}}\\ {r}"))
    display(Math(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}"))
    x0 = round((a + b) / 2, 2)
    x1 = round(x0 - delta_x / 2, 2)
    x2 = round(x0 + delta_x / 2, 2)
    
    display(Math(r"x_0^{" + str(r) + r"} = \frac{a^{" + str(r) + r"} + b^{" + str(r) + r"}}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x0)))
    display(Math(r"x_1^{" + str(r) + r"} = x_0^{" + str(r) + r"} - \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x1)))
    display(Math(r"x_2^{" + str(r) + r"} = x_0^{" + str(r) + r"} + \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x2)))
    
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)
    
    display(Math(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}"))
    display(Math(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}"))
    
    if F1 < F2:
        display(Math(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}"))
        b = x2
    else:
        display(Math(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}"))
        a = x1

    r += 1
display(Math(f"\\textbf{{Ответ:}}\\quad a = {a},\\quad b = {b}"))
```
