# Общие параметры

```python
from IPython.display import display, Math
from math import sqrt

# The function to be changed
def F(x):
    return 2 * (x-12)*(x-12) + 3

def dm(eq):
    display(Math(eq))
eps = 1
```

# Алгоритм золотого сечения

## Начальные параметры

```python
tau = 0.618
a_0 = 0
b_0 = 13
```

## Решение

```python
r = 1
a = a_0
b = b_0
while (b - a) > eps:
    dm(f"\\textbf{{Итерация}}\\ {r}")
    dm(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}")

    x1 = round(b - (b - a) * tau, 2)
    x2 = round(a + (b - a) * tau, 2)
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)

    dm(f"x_1^{{{r}}} = b^{{{r}}} - (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {b} - ({b} - {a}) \\cdot {tau} = {x1}")
    dm(f"x_2^{{{r}}} = a^{{{r}}} + (b^{{{r}}} - a^{{{r}}}) \\cdot \\tau = {a} + ({b} - {a}) \\cdot {tau} = {x2}")

    dm(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}")
    dm(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}")

    if F1 < F2:
        dm(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}")
        b = x2
    else:
        dm(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}")
        a = x1

    r += 1

dm(f"\\textbf{{Ответ:}}\\quad a = {a},\\quad b = {b}")
```

# Алгоритм дихотомического поиска

## Начальные параметры

```python
delta_x = 0.2
```

## Решение

```python
r = 1
a = a_0
b = b_0
while (b - a) > eps:
    dm(f"\\textbf{{Итерация}}\\ {r}")
    dm(f"r = {r},\\quad a^{{{r}}} = {a},\\quad b^{{{r}}} = {b}")
    x0 = round((a + b) / 2, 2)
    x1 = round(x0 - delta_x / 2, 2)
    x2 = round(x0 + delta_x / 2, 2)
    
    dm(r"x_0^{" + str(r) + r"} = \frac{a^{" + str(r) + r"} + b^{" + str(r) + r"}}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x0))
    dm(r"x_1^{" + str(r) + r"} = x_0^{" + str(r) + r"} - \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x1))
    dm(r"x_2^{" + str(r) + r"} = x_0^{" + str(r) + r"} + \frac{\delta_x}{2} = \frac{" + str(a) + " + " + str(b) + "}{2} = " + str(x2))
    
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)
    
    dm(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}")
    dm(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}")
    
    if F1 < F2:
        dm(f"\\Phi(x_1^{{{r}}}) < \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = a^{{{r}}} = {a},\\quad b^{{{r+1}}} = x_2^{{{r}}} = {x2}")
        b = x2
    else:
        dm(f"\\Phi(x_1^{{{r}}}) > \\Phi(x_2^{{{r}}}) \\Rightarrow a^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad b^{{{r+1}}} = b^{{{r}}} = {b}")
        a = x1

    r += 1
dm(f"\\textbf{{Ответ:}}\\quad a = {a},\\quad b = {b}")
```

# Метод квадратичной аппроксимации

## Начальные параметры

Надо условится, что у нас уже имеется функция (с этим проблем нет) и три точки

```python
x1 = 5
x2 = 9
x3 = 13
```

## Решение

```python
a = min(x1, x2, x3)
b = max(x1, x2, x3)

r = 1
while ((x3 - x1) > eps):
    dm(f"\\textbf{{Итерация}}\\ {r}")
    dm(f"r = {r},\\quad x_1^{{{r}}} = {a},\\quad x_2^{{{r}}} = {x2},\\quad x_3^{{{r}}} = {b},\\quad ТИН_{{{r}}} = [x_1^{{{r}}}, x_3^{{{r}}}]")
    
    F1 = round(F(x1), 2)
    F2 = round(F(x2), 2)
    F3 = round(F(x3), 2)
    dm(f"\\Phi(x_1^{{{r}}}) = \\Phi({x1}) = {F1}")
    dm(f"\\Phi(x_2^{{{r}}}) = \\Phi({x2}) = {F2}")
    dm(f"\\Phi(x_3^{{{r}}}) = \\Phi({x3}) = {F3}")

    e23 = round(x2**2 - x3**2, 2)
    e13 = round(x1**2 - x3**2, 2)
    e12 = round(x1**2 - x2**2, 2)

    dm(f"e_{{23}} = x_2^{{2}} - x_3^{{2}} = {round(x2**2, 2)} - {round(x3**2, 2)} = {e23}")
    dm(f"e_{{13}} = x_1^{{2}} - x_3^{{2}} = {round(x1**2, 2)} - {round(x3**2, 2)} = {e13}")
    dm(f"e_{{12}} = x_1^{{2}} - x_2^{{2}} = {round(x1**2, 2)} - {round(x2**2, 2)} = {e12}")

    r23 = round(x2 - x3, 2)
    r13 = round(x1 - x3, 2)
    r12 = round(x1 - x2, 2)

    dm(f"r_{{23}} = x_2 - x_3 = {round(x2, 2)} - {round(x3, 2)} = {r23}")
    dm(f"r_{{13}} = x_1 - x_3 = {round(x1, 2)} - {round(x3, 2)} = {r13}")
    dm(f"r_{{12}} = x_1 - x_2 = {round(x1, 2)} - {round(x2, 2)} = {r12}")
    
    x_tild = round((F1*e23 - F2*e13 + F3*e12)/(2*(F1*r23 - F2*r13 + F3*r12)), 2)
    dm(f"\\tilde{{x}}^{r} = \\frac{{\\Phi_1 \\cdot e_{{23}} - \\Phi_2 \\cdot e_{{13}} + \\Phi_3 \\cdot e_{{12}}}}{{2 \\cdot (\\Phi_1 \\cdot r_{{23}} - \\Phi_2 \\cdot r_{{13}} + \\Phi_3 \\cdot r_{{12}})}} =")
    dm(f"\\frac{{{F1} \\cdot {e23} - {F2} \\cdot {e13} + {F3} \\cdot {e12}}}"
             f"{{2 \\cdot ({F1} \\cdot {r23} - {F2} \\cdot {r13} + {F3} \\cdot {r12})}} = {x_tild}")

    #F_tild = round(F(x_tild), 4)
    if (x2 <= x_tild <= x3):
        dm(f"\\tilde{{x}}^{r} \\in [x_2^{r}, x_3^{r}] \\Rightarrow x_1^{{{r+1}}} = x_2^{{{r}}} = {x2},\\quad x_2^{{{r+1}}} = \\tilde{{x}}^{{{r}}} = {x_tild},\\quad x_3^{{{r+1}}} = x_3^{{{r}}} = {x3}")
        x1 = x2
        x2 = x_tild
    elif (x1 <= x_tild <= x2):
        dm(f"\\tilde{{x}}^{r} \\in [x_1^{r}, x_2^{r}] \\Rightarrow x_1^{{{r+1}}} = x_1^{{{r}}} = {x1},\\quad x_2^{{{r+1}}} = \\tilde{{x}}^{{{r}}} = {x_tild},\\quad x_3^{{{r+1}}} = x_2^{{{r}}} = {x2}")
        x3 = x2
        x2 = x_tild
    dm(f"ТИН_{{{r+1}}} = [x_1^{{{r+1}}}, x_3^{{{r+1}}}] = [{x1}, {x3}]")
    r+=1
    
```

# Алгоритм Фибоначчи

## n-ый элемент ряда Фибоначчи

```python
def n_fib(n):
    τ = (sqrt(5)-1)/2
    return round(((1/τ)**(n+1)-((-τ)**(n+1)))/sqrt(5))
```

## Решение

```python
a=a_0
b=b_0
N = 1
while (b-a)/n_fib(N) > eps:
    N += 1
dm(rf"N={N}")

print("Формулы:")
dm(r"x_{1}^{r}=a^r+|ТИН_r|\frac{i_{N-1-r}}{i_{N+1-r}}")
dm(r"x_{2}^{r}=a^r+|ТИН_r|\frac{i_{N-r}}{i_{N+1-r}}")
print("\n\n") # 3 empty lines for identation

for r in range(1, N): # [1; N-1]
    print(rf"Итерация {r} из {N-1}: ")
    fib1 = n_fib(N-1-r) / n_fib(N+1-r)
    fib2 = n_fib(N-r) / n_fib(N+1-r)
    x1_r = a + abs(b - a) * fib1
    x2_r = a + abs(b - a) * fib2
    Phi1 = F(x1_r)
    Phi2 = F(x2_r)
    
    dm(rf"a_{r} = {a};\ b_{r}={b}")
    dm(rf"\frac{{i_{N-1-r}}}{{i_{N+1-r}}}=\frac{{{n_fib(N-1-r)}}}{{{n_fib(N+1-r)}}};\ \frac{{i_{N-r}}}{{i_{N+1-r}}}=\frac{{{n_fib(N-r)}}}{{{n_fib(N+1-r)}}}")
    dm(rf"x_{1}^{r}={a}+|{b}-{a}|\frac{{i_{N-1-r}}}{{i_{N+1-r}}}={x1_r}")
    dm(rf"x_{2}^{r}={a}+|{b}-{a}|\frac{{i_{N-r}}}{{i_{N+1-r}}}={x2_r}")
    dm(rf"\Phi{{x_1^{r}}}={Phi1}")
    dm(rf"\Phi{{x_2^{r}}}={Phi2}")

    if r != N - 1:
        if Phi1 < Phi2:
            b = x2_r
            dm(rf"\Phi{{x_1^{r}}} < \Phi{{x_2^{r}}} \Rightarrow b_{r+1} = x_2^{r} = {x2_r}")
        else:
            a = x1_r
            dm(rf"\Phi{{x_1^{r}}} > \Phi{{x_2^{r}}} \Rightarrow a_{r+1} = x_1^{r} = {x1_r}")
    
    print("\n") # 2 empty lines for identation

assert x1_r == x2_r
print("Этап 2")
r += 1

delta_x = (b - a)/20
x2_r = x1_r + delta_x
Phi1 = F(x1_r)
Phi2 = F(x2_r)

dm(rf"a_{r} = {a};\ b_{r}={b}")
dm(rf"\text{{Пусть }}\delta_x=\frac{{|ТИН_r|}}{{20}}={delta_x}")
dm(rf"x_{1}^{r} = {x1_r}")
dm(rf"x_{2}^{r} = x_{1}^{r-1} + \delta_x = {x2_r}")
dm(rf"\Phi{{x_1^{r}}}={Phi1}")
dm(rf"\Phi{{x_2^{r}}}={Phi2}")

if Phi1 < Phi2:
    b = x1_r
    dm(rf"\Phi{{x_1^{r}}} < \Phi{{x_2^{r}}} \Rightarrow b_{r} = x_1^{r} = {x1_r}")
else:
    a = x1_r
    dm(rf"\Phi{{x_1^{r}}} > \Phi{{x_2^{r}}} \Rightarrow a_{r} = x_1^{r} = {x1_r}")
dm(rf"a_{r} = {a};\ b_{r}={b}")
```
